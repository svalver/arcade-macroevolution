```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({'font.size': 15})
plt.rcParams['svg.fonttype'] = 'none'
# matplotlib_inline.backend_inline.set_matplotlib_formats("svg")
%config InlineBackend.figure_formats = ['svg'] 
```

```python

# Parameters
a = 1.44  # scaling exponent for S ~ M^a
k = 0.1   # scaling constant

# Memory range (M)
M = np.linspace(1, 10, 100)
S = k * M**a

# Generate "points" around the curve for high and low b regimes
np.random.seed(42)

# High-diversity (left panel)
M_high = np.random.uniform(1, 10, 150)
S_high = k * M_high**a + np.random.normal(0, 0.5, size=M_high.shape)  # wide spread

# Low-diversity (right panel)
M_low = np.random.uniform(1, 10, 100)
S_low = k * M_low**a + np.random.normal(0, 0.1, size=M_low.shape)  # tight spread

# Create figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5), sharey=True)
# fig.suptitle("Figure C.1: Constraint-Induced Canalization in Arcade Morphospace", fontsize=14)

# Left: High-diversity
ax1.plot(M, S, 'k--', label='$S \\sim M^a$')
ax1.scatter(M_high, S_high, alpha=0.5, label='Niche trajectories')
ax1.set_title("High diversity (b ≈ 1)")
ax1.set_xlabel("ROM Size (M)")
ax1.set_ylabel("CPU Speed (S)")
# ax1.legend()

# Right: Low-diversity
ax2.plot(M, S, 'k--', label='$S \\sim M^a$')
ax2.scatter(M_low, S_low, alpha=0.5, color='darkred', label='Niche trajectories')
ax2.set_title("Collapsed niche (b ≪ 1)")
ax2.set_xlabel("Rom Size (M)")
# ax2.legend()

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig ("fig_c1.pdf", bbox_inches='tight', dpi=70)
plt.show()
```
